```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
from selenium import webdriver
import requests
```

```python
url = 'https://www.baseball-reference.com/leagues/majors/2025-standard-batting.shtml'
driver = webdriver.Chrome()
driver.get(url)
html = driver.page_source
soup = BeautifulSoup(html)
table = soup.find('table', id='players_standard_batting')
tbody = table.find('tbody')

players = []
names = []
for p in tbody.find_all('tr'):
    if p.get('class') == None or 'thead' not in p.get('class'):
        players.append(p)
        for td in p.find_all('td', attrs={'data-stat':'name_display'}):
            if td.a is not None:
                names.append(td.a.get_text(strip=True))
    else:
        continue

for n in names:
    print(n)

print(len(players))
print(len(names))
```
